```python
from datasets import load_dataset

ds = load_dataset("agie-ai/Open-Orca-OpenOrca",split="train",streaming=True)
```

```python
testData = list(small_dataset)
datasetTEMP = Dataset.from_list(testData)
```

```python
df = datasetTEMP.to_pandas()
```
