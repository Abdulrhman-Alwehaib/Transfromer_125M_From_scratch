```python
from datasets import load_dataset
from datasets import Dataset
import pandas as pd
```

```python
datasetsCount = 1
datasetsLINKS = ["ClusterlabAi/101_billion_arabic_words_dataset"]
datasetLINKSFeature = ["text"]
files = "data/train-0000[0-2]-of-*.parquet"
for i in range(datasetsCount):
    dataset = load_dataset(datasetsLINKS[i],
                           data_files=files,
                           verification_mode="no_checks",
                           )
    trainingSpecfic = dataset['train']
    datasetTEMP = dataset.select_columns([datasetLINKSFeature[i]])
    df = trainingSpecfic.to_pandas()
    df.to_csv(f"../Data/RAW_dataset/ArabicLLM{i + 1}",index=False,escapechar='\\')
```
